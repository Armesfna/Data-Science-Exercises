# <font color='red'> <b> Why Folium</b><font color='red'>

Folium is a map visualization library developed for the Python programming language. It leverages the mapping visualization functionality using the JavaScript-based Leaflet.js library in the background. With this library, we can easily process map data and create interactive maps

One interesting thing about Folium maps is that they're interactive. This means that even if you set a specific zoom level at the beginning, you can zoom in on any area you're interested in to see it more clearly.

```python
# Import Folium libraries
import folium

# Adjust the coordinates of Paris
paris_coordinate = (48.8566, 2.3522)

# Create a Map
paris_map = folium.Map(location=paris_coordinate, zoom_start=11)

# Add markers to the Eiffel Tower
folium.Marker(location=(48.8584, 2.3240),
              popup='Eyfel Kulesi',
              icon=folium.Icon(color='red')).add_to(paris_map)

# Add markers to other tourist attractions
tourist_places = {
    "Louvre Müzesi": (48.8629, 2.3358),
    "Notre Dame Katedrali": (48.8651, 2.3450),
    "Zafer Takı": (48.8736, 2.2762),
    "Montmartre Tepesi": (48.8844, 2.3423),
    "Sacré-Cœur Bazilikası": (48.8860, 2.3419),
}

for ad, coordinate in tourist_places.items():
    folium.Marker(location=coordinate, popup=ad).add_to(paris_map)

# puts a link to the location of the project in html
# paris_haritasi.save("paris_haritasi.html")
# Show Map
paris_map
```

Some values the tiles parameter can take are:

CartoDB Positron: This offers a modern and minimalist map style. It has grey tones, white labels, and dark borders, making it good for data visualizations.

Stamen Terrain: This is a colorful map style showing natural landforms and topographic features. It highlights mountains, hills, bodies of water, and green areas.

Stamen Watercolor: This has a hand-drawn style in watercolors. It highlights natural features like water, land, and vegetation, with a high artistic design.

Stamen Toner: This is a black and white map style. It emphasizes details like roads, borders, and labels. It is ideal for data visualizations.

CartoDB Dark Matter: This has a black background with white labels, offering a minimalist map style. It is suitable for scientific maps, visualizations, and analyses.

```python
import folium

# Coordinates and information of historical monuments in Istanbul
istanbul_historical_monuments = {
    "Hagia Sophia": {"coordinates": [41.0084, 28.9799], "built_year": "537"},
    "Topkapi Palace": {"coordinates": [41.0115, 28.9833], "built_year": "1478"},
    "Suleymaniye Mosque": {"coordinates": [41.0165, 28.9584], "built_year": "1557"},
    "Basilica Cistern": {"coordinates": [41.0085, 28.9799], "built_year": "6th century"},
    "Galata Tower": {"coordinates": [41.0257, 28.9744], "built_year": "1348"}
}

# Create a map centered at a point in Istanbul
m = folium.Map(location=[41.0082, 28.9784], zoom_start=13)

# Add a marker for each historical monument
for monument, info in istanbul_historical_monuments.items():
    coordinates = info["coordinates"]
    built_year = info["built_year"]
    popup_text = f'<strong>{monument}</strong><br>Built Year: {built_year}'
    tooltip_text = f'{monument} ({built_year})'
    folium.Marker(
        location=coordinates,
        popup=popup_text,
        tooltip=tooltip_text
    ).add_to(m)

# Display the map
m
```

<a id="displaying-line-between-2-distances"></a>
## Displaying Line Between 2 Distances

```python
# Import the required Python libraries
import folium  # This library helps create interactive maps
from geopy.distance import geodesic  # This library calculates geographical distances

# Create a folium map centered around the given latitude and longitude with a certain zoom level
m = folium.Map([31.335432273991238, 75.53764505992153], zoom_start=6)

# Add a marker to the map with a specific location and a popup message
m.add_child(folium.Marker(location=[31.333, 75.537], popup="First Marker"))

# Create a feature group to organize map features
fg = folium.FeatureGroup(name="My map")

# Add a marker with a green color icon to the feature group
fg.add_child(
    folium.Marker(location=[32.00, 75.00],
                  popup="2nd",
                  icon=folium.Icon(color="green"))
)

# Add the feature group to the map
m.add_child(fg)

# Calculate the distance between two points in miles and convert it to kilometers
distance_miles = geodesic([31.333, 75.537], [32.00, 75.00]).miles
distance_km = distance_miles * 1.60934

# Create a polygon between two points on the map and display the calculated distance as a popup
folium.Polygon(locations=[[31.333, 75.537], [32.00, 75.00]],
               popup=f"The distance is approximately {distance_km:.2f} kilometers.",
               color='blue').add_to(m)

# Display the map
m
```

<a id="location-search-feature"></a>
## Location Search Feature

The Geocoder plugin is added to the map to enable location search functionality.
- collapsed parameter determines whether the search box is open or closed. If set to 
- False, the search box is open by default.
-  position parameter sets the position of the search box on the map. It can be set to values like "topleft", "topright", "bottomleft", "bottomright".
- add_marker parameter determines whether search results are marked on the map. If set to True, the search results are marked.

```python
# Import Folium libraries
import folium

# Create the map
m = folium.Map(location=[41.015, 28.979], zoom_start=13)

# Add Geocoder plugin to enable location search on the map
folium.plugins.Geocoder(collapsed=False, position="topleft", add_marker=True).add_to(m)

# Display the map
m
```

## Full-Screen View of the Map

The FullscreenControl plugin is added to the map to enable fullscreen functionality. It adds a button on the map that allows users to view the map in fullscreen mode.
position parameter sets the position of the fullscreen button on the map. It is set to 'topright', meaning the button will appear at the top right corner of the map.

```python
import folium
from folium import plugins

# Create the map
istanbul_map = folium.Map(location=[41.015137, 28.979530], zoom_start=12)

# Add Fullscreen plugin
plugins.Fullscreen().add_to(istanbul_map)

# Display the map
istanbul_map
```

Add Fullscreen Plugin: We add the Fullscreen plugin to the map using the plugins.Fullscreen().add_to() method. This plugin adds a button to the map interface, allowing users to view the map in fullscreen mode.

# <font color='red'> <b>In which areas are our customers concentrated in New York?</b><font color='red'>

```python
#  Turn customer data into a list
customers = [
    {"customer_id": 1, "region": "Queens", "enlem": 41.4036, "boylam": 2.1743},
    {"customer_id": 2, "region": "Manhattan", "enlem": 40.7128, "boylam": -74.0060},
    {"customer_id": 3, "region": "Brooklyn", "enlem": 40.6781, "boylam": -73.9756},
    {"customer_id": 4, "region": "Queens", "enlem": 41.4036, "boylam": 2.1606},
    {"customer_id": 5, "region": "Manhattan", "enlem": 40.7128, "boylam": -74.0060},
    {"customer_id": 6, "region": "Bronx", "enlem": 40.8448, "boylam": -73.8648},  # Eklediğiniz yeni müşteri
    {"customer_id": 7, "region": "Staten Island", "enlem": 40.5795, "boylam": -74.1502}  # Eklediğiniz yeni müşteri
]

# Import Folium libraries
import folium

# Create a New York map
new_york_map = folium.Map(location=[40.7128, -74.0060], zoom_start=10)

# Dictionary of region coordinates
region_coordinates = {
    "Queens": (40.7282, -73.7949),
    "Manhattan": (40.7831, -73.9712),
    "Brooklyn": (40.6782, -73.9442),
    "Bronx": (40.8448, -73.8648),  
    "Staten Island": (40.5795, -74.1502) 
}

# Add customer data to the map
for customer in customers:
    # Gracefully handle missing regions (Option 1)
    location = region_coordinates.get(customer["region"])
    if location:
        folium.Marker(location=location, popup=f"{customer['customer_id']} - {customer['region']}").add_to(new_york_map)
    else:
        # Add a default marker for missing regions (optional)
        default_location = (40.7128, -74.0060)  # New York City center (example)
        folium.Marker(location=default_location, popup=f"{customer['customer_id']} - **{customer['region']}**").add_to(new_york_map)

# Show the map
new_york_map
```

<a id="map-amazon-top-5-selling-countries"></a>
# <font color='red'> <b>Map of Amazon's Top 5 Selling Countries</b><font color='red'>

```python
# Convert Amazon sales data to a list
amazon_sales = [
    {"country": "US", "sales_amount": 100000000},
    {"country": "DE", "sales_amount": 50000000},
    {"country": "GB", "sales_amount": 40000000},
    {"country": "JP", "sales_amount": 30000000},
    {"country": "FR", "sales_amount": 20000000},
]

import folium

# Create the world map
world_map = folium.Map(location=[0, 0], zoom_start=2)

# Country coordinates
country_coordinates = {
    "US": [37.0902, -95.7129],
    "DE": [51.1657, 10.4515],
    "GB": [55.3781, -3.4360],
    "JP": [36.2048, 138.2529],
    "FR": [46.6034, 1.8883]
}

# Find the country with the highest sales
top_selling_country, top_sales_amount = max([(sale["country"], sale["sales_amount"]) for sale in amazon_sales], key=lambda x: x[1])

# Add markers to the map
for sale in amazon_sales:
    if sale["country"] in country_coordinates:
        popup = f"{sale['country']} - Sales Amount: {sale['sales_amount']:,}"
        folium.Marker(location=country_coordinates[sale["country"]], popup=popup).add_to(world_map)

# Highlight the marker of the top selling country in a different color
if top_selling_country in country_coordinates:
    popup = f"**Top Sales:** {top_selling_country} - {top_sales_amount:,}"
    folium.Marker(location=country_coordinates[top_selling_country], popup=popup, icon=folium.Icon(color='red')).add_to(world_map)

# Save the map
#world_map.save("amazon_sales_map.html")

world_map
```

# <font color='red'> <b>The 5 Most Populous Countries by Population Density</b><font color='red'>

```python
import folium

# Population data
population_data = [
    {"country": "China", "population": 1444215628},
    {"country": "India", "population": 1380004385},
    {"country": "United States", "population": 332975285},
    {"country": "Indonesia", "population": 273523615},
    {"country": "Pakistan", "population": 235309000},
]

# Coordinates of selected countries
country_coordinates = {
    "China": {"latitude": 35.8617, "longitude": 104.1954},
    "India": {"latitude": 20.5937, "longitude": 78.9629},
    "United States": {"latitude": 37.0902, "longitude": -95.7129},
    "Indonesia": {"latitude": -0.7893, "longitude": 113.9213},
    "Pakistan": {"latitude": 30.3753, "longitude": 69.3451}
}

# Create the world map
world_map = folium.Map(location=[0, 0], zoom_start=2)

# Find the most populous country
most_populous_country = max(population_data, key=lambda x: x["population"])

# Add markers to the map
for country_data in population_data:
    country = country_data["country"]
    popup = f"{country} - Population: {country_data['population']:,}"
    color = 'red' if country_data == most_populous_country else 'blue'
    if country in country_coordinates:
        coordinates = country_coordinates[country]
        folium.Marker(location=[coordinates["latitude"], coordinates["longitude"]], popup=popup, icon=folium.Icon(color=color)).add_to(world_map)

# Show map
world_map
```
