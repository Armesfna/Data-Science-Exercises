```python
import matplotlib.pyplot as plt

# Days of the week
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']

# Sales data for different products
products = {
    'Apples': [20, 25, 30, 35, 40],
    'Oranges': [15, 20, 25, 30, 35],
    'Bananas': [10, 12, 15, 18, 20]
}

# Extracting product names and quantities into separate lists
product_names = list(products.keys())
quantities = [list(products[product]) for product in product_names]

# Colors for each product
colors = ['#ff9999', '#66b3ff', '#99ff99']

# Creating the stackplot with customized colors and labels
plt.stackplot(days, quantities, labels=product_names, colors=colors)

# Adding labels and title
plt.xlabel('Days of the Week')
plt.ylabel('Quantity Sold')
plt.title('Weekly Sales of Fruits')

# Adding legend with customized location
plt.legend(loc='upper left')

# Customizing tick marks and grid lines
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adding shadow effect to the legend box
plt.legend(loc='upper left', shadow=True)

# Displaying the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Months of the year
months = np.arange(1, 13)

# Monthly sales data for furniture products (sample data)
furniture_sales = {
    'Sofa': [5000, 5200, 5400, 5600, 5800, 6000, 6200, 6400, 6600, 6800, 7000, 7200],
    'Bed': [4000, 4200, 4400, 4600, 4800, 5000, 5200, 5400, 5600, 5800, 6000, 6200],
    'Desk': [3000, 3200, 3400, 3600, 3800, 4000, 4200, 4400, 4600, 4800, 5000, 5200]
}

# Separate product names and sales quantities into lists
product_names = list(furniture_sales.keys())
quantities = [list(furniture_sales[product]) for product in product_names]

# Custom colors
colors = ['#ff9999', '#66b3ff', '#99ff99']

# Create stackplot and add custom colors and labels
plt.stackplot(months, quantities, labels=product_names, colors=colors)

# Add axis labels and title
plt.xlabel('Months')
plt.ylabel('Sales Quantity')
plt.title('Monthly Sales of Furniture Products')

# Add legend and position it
plt.legend(loc='upper left')

# Add gridlines and shaded legend box
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(loc='upper left', shadow=True)

# Display the plot
plt.tight_layout()
plt.show()
```
