```python
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv('data_05_Filling_Area_on_Line_Plots.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

plt.fill_between(ages, py_salaries, dev_salaries,
                 where=(py_salaries > dev_salaries),
                 interpolate=True, alpha=0.25, label='Above Avg')

plt.fill_between(ages, py_salaries, dev_salaries,
                 where=(py_salaries <= dev_salaries),
                 interpolate=True, color='red', alpha=0.25, label='Below Avg')

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Monthly sales data
months = np.arange(1, 13)
sales = [20000, 22000, 25000, 28000, 30000, 35000, 38000, 40000, 42000, 45000, 48000, 50000]

# Create a figure and axis object
fig, ax = plt.subplots()

# Plot the monthly sales
ax.plot(months, sales, marker='o', linestyle='-', color='blue', label='Monthly Sales')

# Fill the area under the curve where sales exceed 30000 with a light blue color
ax.fill_between(months, sales, where=(np.array(sales) > 30000), color='lightblue', alpha=0.8)

# Add gridlines
ax.grid(True, linestyle='--', alpha=0.7)

# Add title and labels
ax.set_title('Monthly Sales Performance')
ax.set_xlabel('Months')
ax.set_ylabel('Sales')

# Add legend
ax.legend()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Countries and monthly bread consumption (example data)
countries = ['USA', 'Germany', 'France', 'Italy', 'Turkey']
bread_consumption = [200, 180, 160, 220, 240]  # Monthly bread consumption in kilograms

# Create a figure and axis object
fig, ax = plt.subplots()

# Plot a line graph showing the bread consumption of each country
ax.plot(countries, bread_consumption, marker='o', linestyle='-', color='blue', label='Monthly Bread Consumption')

# Fill the area under the line where bread consumption is over 200, in light blue
ax.fill_between(countries, bread_consumption, where=(np.array(bread_consumption) > 200), color='lightblue', alpha=0.6)

# Add gridlines
ax.grid(True, linestyle='--', alpha=0.7)

# Add title and labels
ax.set_title('Monthly Bread Consumption by Country')
ax.set_xlabel('Countries')
ax.set_ylabel('Bread Consumption (kg)')

# Add legend
ax.legend()

# Show the plot
plt.show()
```
